```python
import numpy as np

reduction = lambda N, X:  (1/1000) * 15 * (10/51)**N / X

X = 51


N = np.array([i for i in range(0,10)])

import matplotlib.pyplot as plt

plt.plot(N, reduction(N, X))
plt.yscale('log')
plt.xlabel('Engranajes N')
plt.grid()
plt.show()

```

```python
import numpy as np

reduction = lambda N, X:  (1/300) * 15 * (10/51)**N / X

X = 51

N = np.array([i for i in range(0,10)])

import matplotlib.pyplot as plt
for x in [51, 30, 15]:
    plt.plot(N, (360/2**12)* reduction(N, x), label=f'x={x:.2f}') 
    plt.yscale('log')
    plt.xlabel('Engranajes N')
    plt.title('Step resultante')
    plt.legend()
    plt.grid()
plt.show()

```

```python
import numpy as np

reduction = lambda N, X:  (1/300) * 15 * (10/51)**N / X

X = 51
RPM = 13

N = np.array([i for i in range(0,10)])

import matplotlib.pyplot as plt
for x in [51, 30, 15]:
    plt.plot(N, (RPM)* reduction(N, x), label=f'x={x:.2f}') 
    plt.yscale('log')
    plt.xlabel('Engranajes N')
    plt.title('RPM resultante')
    plt.legend()
    plt.grid()
plt.show()

```
